```python
import pandas as pd
```

```python
df = pd.read_csv('data/netflow/CIDDS-001/traffic/mixed.csv', index_col='DATE')
df.head()
```

```python
import numpy as np
```

```python
full_true_set = df.sample(20)
```

```python
full_true_set
```

```python
full_pred_set = df.sample(20)
```

```python
full_pred_set
```

```python
true_set = full_true_set['ATTACK_TYPE'].tolist()
print(true_set)
pred_set = full_pred_set['ATTACK_TYPE'].tolist()
print(pred_set)
```

```python
from sklearn import metrics
```

```python
print("Confusion matrix")

conf_mat = metrics.confusion_matrix(true_set, pred_set, labels=['benign', 'dos', 'portScan', 'pingScan', 'bruteForce'])
print(conf_mat)
labels = ['benign', 'dos', 'portScan', 'pingScan', 'bruteForce']
```

```python
# TODO: it may be better to direcly use plot_confusion_matrix, but it needs a Classifier. As we don't have one yet we use another mean to plot the matrix.
#       when the classifier will be ready, we will be able to use this function directly to produce both y_pred and the plot.

import seaborn as sn
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16,12))
sn.heatmap(conf_mat, annot=True, ax = ax);

ax.set_xlabel('Predicted labels');ax.set_ylabel('True labels'); 
ax.set_title('Confusion Matrix'); 
ax.xaxis.set_ticklabels(['benign', 'dos', 'portScan', 'pingScan', 'bruteForce']);
ax.yaxis.set_ticklabels(['bruteForce', 'pingScan', 'portScan', 'dos', 'benign']);
```

```python
print("Classification accuracy percentage") # Classification rate
metrics.accuracy_score(true_set, pred_set) * 100
```

```python
print("Classification error percentage")
(1 - metrics.accuracy_score(true_set, pred_set)) * 100
```

```python
print("Sensitivity") # Detection rate or Recall or True positive rate
print(metrics.recall_score(true_set, pred_set, average="micro"))
```

```python
print("Precision")
print(metrics.precision_score(true_set, pred_set, average="micro"))
```
