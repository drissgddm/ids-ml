```python
from sklearn.preprocessing import LabelEncoder

def preprocess_port_feature(series_to_encode, vals_to_encode=[80, 53, 443, 3389, 445, 21, 8080, 22, 53]):
    lb_e = LabelEncoder()
    series_to_encode = series_to_encode.where(series_to_encode.isin(vals_to_encode))
    series_to_encode = lb_e.fit_transform(series_to_encode)
    return series_to_encode    
```

```python
df['SRC_PORT'] = preprocess_port_feature(df['SRC_PORT'])
df['DST_PORT'] = preprocess_port_feature(df['DST_PORT'])
```

```python
flags_cols = ['FLAG_A', 
    'FLAG_P',
    'FLAG_R',
    'FLAG_S',
    'FLAG_F',
    'FLAG_0xdb',
    'FLAG_0xc2',
    'FLAG_0xd7',
    'FLAG_0xd6',
    'FLAG_0xdf',
    'FLAG_0xda',
    'FLAG_0x52',
    'FLAG_0xd3',
    'FLAG_0x5b',
    'FLAG_0x5a',
    'FLAG_0xd2',
    'FLAG_0xde',
    'FLAG_0x53',
    'FLAG_0xc6']
```

```python
import numpy as np
```

```python
def preprocess_flags_feature(df):
    
    def apply_custom_one_hot_encoding(flag_val):
        r, flags_map = {
            'FLAG_A': 0,
            'FLAG_P': 0,
            'FLAG_R': 0,
            'FLAG_S': 0,
            'FLAG_F': 0,
            'FLAG_0xdb': 0,
            'FLAG_0xc2': 0, 
            'FLAG_0xd7': 0,
            'FLAG_0xd6': 0,
            'FLAG_0xdf': 0,
            'FLAG_0xda': 0,
            'FLAG_0x52': 0,
            'FLAG_0xd3': 0,
            'FLAG_0x5b': 0,
            'FLAG_0x5a': 0,
            'FLAG_0xd2': 0,
            'FLAG_0xde': 0,
            'FLAG_0x53': 0,
            'FLAG_0xc6': 0
        }, {
            '.A....': ['FLAG_A'], 
            '.AP.SF': ['FLAG_A', 'FLAG_P', 'FLAG_S', 'FLAG_F'], 
            '.AP...': ['FLAG_A', 'FLAG_P'], 
            '.AP.S.': ['FLAG_A', 'FLAG_P', 'FLAG_S'], 
            '.A..SF': ['FLAG_A', 'FLAG_S', 'FLAG_F'],
            '.A...F': ['FLAG_A', 'FLAG_F'],
            '.A..S.': ['FLAG_A', 'FLAG_P', 'FLAG_F'],
            '.AP..F': ['FLAG_A', 'FLAG_P', 'FLAG_F'],
            '....S.': ['FLAG_S'], 
            '.A.R..': ['FLAG_A', 'FLAG_R'],
            '.APRSF': ['FLAG_A', 'FLAG_P', 'FLAG_R', 'FLAG_S', 'FLAG_F'], 
            '...R..': ['FLAG_R'],
            '.APR.F': ['FLAG_A', 'FLAG_P', 'FLAG_R' 'FLAG_F'],
            '...RS.': ['FLAG_R', 'FLAG_S'],
            '.APRS.': ['FLAG_A', 'FLAG_P', 'FLAG_R' 'FLAG_S'],
            '.A.R.F': ['FLAG_A', 'FLAG_R' 'FLAG_F'],
            '.APR..': ['FLAG_A', 'FLAG_P', 'FLAG_R'],
            '.A.RSF': ['FLAG_A', 'FLAG_R', 'FLAG_S' 'FLAG_F'],
            '.A.RS.': ['FLAG_A', 'FLAG_R', 'FLAG_S'],
            '  0xdb': ['FLAG_0xdb'],
            '  0xc2': ['FLAG_0xc2'],
            '  0xd7': ['FLAG_0xd7'], 
            '  0xd6': ['FLAG_0xd6'], 
            '  0xdf': ['FLAG_0xdf'], 
            '  0xda': ['FLAG_0xda'], 
            '  0x52': ['FLAG_0x52'], 
            '  0xd3': ['FLAG_0xd3'],
            '  0x5b': ['FLAG_0x5b'], 
            '  0x5a': ['FLAG_0x5a'], 
            '  0xd2': ['FLAG_0xd2'], 
            '  0xde': ['FLAG_0xde'], 
            '  0x53': ['FLAG_0x53'], 
            '  0xc6': ['FLAG_0xc6']
        }

        features = flags_map.get(flag_val.rstrip())
        if features:
            for f in features:
                r[f] = 1
        return r
    
    return pd.DataFrame([apply_custom_one_hot_encoding(row['FLAGS']) for _, row in df.iterrows()])
```

```python
flags_df = preprocess_flags_feature(df)
flags_df.index = df.index
```

```python
FEATURE_TO_LABELENCODE = ['PROTOCOL', 'TYPE', 'SRC_IP_ADDR', 'DST_IP_ADDR', 'TOS', 'ATTACK_TYPE']
```

```python
def label_encode(series_to_encode):
    lb_e = LabelEncoder()
    series_to_encode = lb_e.fit_transform(series_to_encode)
    return series_to_encode
```

```python
for f in FEATURE_TO_LABELENCODE:
    print("Processing feature {}".format(f))
    df[f] = label_encode(df[f])
```
